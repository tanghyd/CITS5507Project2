```python
print('pivots')
world_size = 8
depth = int(np.log2(world_size))
for rank in range(world_size):
#     print(f'rank: {rank}')
    for current_depth in range(depth):
        partition_size = world_size // (2**current_depth)
        pivot_source = int((rank // partition_size) * partition_size)
        
        if rank == pivot_source:
            destination = [k + rank for k in range(partition_size)]
            print(f'depth: {current_depth} | source: {rank} | destination: {destination}')
#         print(f'depth: {current_depth}: partition_size: {partition_size} pivot_source: {pivot_source}')
```

```python
print('partitions')
world_size = 8
depth = int(np.log2(world_size))
for rank in range(world_size):
    for current_depth in range(depth):
        partition_size = world_size // (2**current_depth)
        
        if rank % 2 == 0:  # even
            print(f'depth: {current_depth} | source: {rank} | destination: {rank + 1}')
        else:  # odd
            print(f'depth: {current_depth} | source: {rank} | destination: {rank - 1}')
```
